```python
def do_arithmetic( x, y, op ='add' ):
    
    """ 
    Performs a given operation (third argument) between two given numbers (float or integer).

    Parameters:
    ----------
    x: The first number.
    Type x: integer or float.
    
    y: The second number.
    Type y: integer or float.
    
    op: One of the four operations: 'add' for addition, 'substract' for subtraction,
    'multiply' for multiplication and 'divide' for division with default operation the addition
    between the first two arguments x,y.
    Type op: String.
    
    Returns:
    ---------
    Returns the result of the operation between the given two numbers.
    Return type: float or None if the operation is not valid.
    
    """
    # Clarify the four valid operations in a list.
    Operations=['add','multiply','subtract','divide' ]
    
    #Lowercase the given operation to match the strings in the Operation List.
    Given_operation = op.lower()
    
    #Calculate the results based on the given operations between the two numbers.
    if Given_operation in Operations:
        
        if Given_operation == 'add':
            z = (x+y)
            
        elif Given_operation == 'multiply':
            z = (x*y)
            
        elif Given_operation == 'subtract':
            z= (x-y)
            
        #Checking the division cases and distinguish if the second argument in the division is equal to zero.
        #If so the operation is not valid and returns None.
    
        elif Given_operation == 'divide':
            if y == 0:
                return None
            
            else:
                z= float(x/ y)
                
    #If the given operation is not in the four valid operations returns 'Unknown operation'
    else:
        print("Unknown operation")
        return None
    return float(z)

def sum_of_digits(s):
    
    """
    Gets one string as input and takes into account only the digits and calculate their sum.
    
    Parameters:
    ----------
    s: A string that can contain letters, numbers or symbols.
    Type: String.
    
    Returns:
    ---------    
    Returns the sum of the numbers from the given string (s).
    Return type: Integer.
    """
    
    import re 
    return (sum(int(x) for x in re.findall(r'[0-9]{1,1}',s)))

    #Or return sum(int(x) for x in s if x.isnumeric())


### --- TEST THE CODE ---
if __name__ == '__main__':
    testcases = {'do_arithmetic': [(10, 4, 'add'), (2, 3, 'multiply')],
    'sum_of_digits':[("123",), ("10a20",)]}

    print('\n-- do_arithmetic testcases --')
    for args in testcases['do_arithmetic']:
        print('input:', str(args))
        print('output:', do_arithmetic(*args))
        print('-----------')

    print('\n-- sum_of_digits testcases --')
    for args in testcases['sum_of_digits']:
        print('input:', str(args))
        print('output:', sum_of_digits(*args))
        print('-----------')
```
